# FinData - Market & ML Analysis

Notebook inicial para explorar métricas do pipeline e desempenho dos modelos baseline.

```python
import json
from pathlib import Path
import pandas as pd
```

```python
metrics_path = Path('..') / 'ml' / 'artifacts' / 'metrics_baseline.json'
metrics = json.loads(metrics_path.read_text(encoding='utf-8'))
metrics
```

```python
classification_df = pd.DataFrame(metrics['classification']['models']).T
classification_df.sort_values('f1', ascending=False)
```

```python
regression_df = pd.DataFrame(metrics['regression']['models']).T
regression_df.sort_values('rmse')
```
